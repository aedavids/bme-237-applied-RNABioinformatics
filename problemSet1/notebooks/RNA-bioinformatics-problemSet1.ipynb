## Question 11
Find a splice junction that was detected by STAR, but not detected by HISAT2.

What is the chromosome, start position, and end position of the intron that corresponds to this splice junction?

```python
!ls featureCounts.out
```

```python
starJunctions   = "featureCounts.out/star.ENCFF000IJ.junctions"
hisat2Junctions = "featureCounts.out/hisat2.ENCFF000IJ.junctions"
```

```python
starJuncDF   = pd.read_csv( starJunctions, sep='\t', skiprows = 1 )
starJuncDF.rename(columns={ starJuncDF.columns[-1]: "count" }, inplace = True)

hisat2JuncDF = pd.read_csv( hisat2Junctions, sep='\t', skiprows = 1 )
hisat2JuncDF.rename(columns={ hisat2JuncDF.columns[-1]: "count" }, inplace = True)
```

```python
sc = ["s-" + c for c in starJuncDF]
hc = ["h-" + c for c in hisat2JuncDF]
df = pd.concat( [starJuncDF, hisat2JuncDF], axis=1)
df.columns = sc + hc
df.head()
```

```python
selectRowsWithReads = ( (df["s-count"] != 0) | (df["h-count"] != 0) )
hasReadsDF = df.loc[ selectRowsWithReads, : ]
hasReadsDF.head()
```

```python
selectStarOnlyRows = ( (hasReadsDF["s-count"] != 0) & (hasReadsDF["h-count"] == 0) )
starOnlyDF = hasReadsDF.loc[ selectStarOnlyRows ]
starOnlyDF.loc[:,["s-Geneid", "s-count", "s-Length", "h-Geneid", "h-count",]].head()
```

```python
starOnlyDF.loc[ starOnlyDF["s-Geneid"] == "ENSG00000284740.1", ["s-Geneid", "s-Chr", "s-Start", "s-End", "s-Length"] ]
```
